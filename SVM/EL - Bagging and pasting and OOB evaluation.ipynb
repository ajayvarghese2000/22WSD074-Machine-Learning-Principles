```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_moons(n_samples=1000, noise=0.15, random_state=42)
```

```python
std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,test_size = 0.50, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
bag_clf = BaggingClassifier(
DecisionTreeClassifier(), n_estimators=500,
max_samples=100, bootstrap=True, n_jobs=-1)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

```python
bag_clf2 = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    bootstrap=True, n_jobs=-1, oob_score=True)

bag_clf2.fit(X_train, y_train)
bag_clf2.oob_score_
```

```python
from sklearn.metrics import accuracy_score
y_pred = bag_clf2.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
bag_clf2.oob_decision_function_
```

```python
# do different training data split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y,test_size = 0.50, random_state=42)

bag_clf = BaggingClassifier(
DecisionTreeClassifier(), n_estimators=500,
max_samples=100, bootstrap=True, n_jobs=-1)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)

bag_clf2 = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    bootstrap=True, n_jobs=-1, oob_score=True)

bag_clf2.fit(X_train, y_train)
bag_clf2.oob_score_
```

```python
from sklearn.metrics import accuracy_score
y_pred = bag_clf2.predict(X_test)
accuracy_score(y_test, y_pred)
```
